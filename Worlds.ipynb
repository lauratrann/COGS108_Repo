```python
# Imports - these are all the imports needed for the assignment
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import patsy
import statsmodels.api as sm
import scipy.stats as stats
from scipy.stats import ttest_ind, chisquare, normaltest
```

```python
df_worlds = pd.read_csv('oct2015.csv')
df_sept2016 = pd.read_csv('sept2016.csv')
df_oct2016 = pd.read_csv('oct2016.csv')
df_sept2017 = pd.read_csv('sept2017.csv')
df_oct2017 = pd.read_csv('oct2017.csv')
df_nov2017 = pd.read_csv('nov2017.csv')
df_oct2018 = pd.read_csv('oct2018.csv')
df_nov2018 = pd.read_csv('nov2018.csv')

df_worlds = df_worlds.append(df_sept2016)
df_worlds = df_worlds.append(df_oct2016)
df_worlds = df_worlds.append(df_sept2017)
df_worlds = df_worlds.append(df_oct2017)
df_worlds = df_worlds.append(df_nov2017)
df_worlds = df_worlds.append(df_oct2018)
df_worlds = df_worlds.append(df_nov2018)

print(df_worlds)
```

```python
df_worlds = df_worlds.drop(columns=['Unnamed: 0', 'Followers per hour', 'Followers gained','Games'], axis = 1)
print(df_worlds)
```

```python
### dropped the outliers where views per hour = 0 
df_worlds = df_worlds[(df_worlds['Views per hour'] > 0)]
df_worlds
```

```python
#Followed Leo's cleaning and also will convert minutes streamed to hours for consistency
def change_hours(inValue):
    return inValue / 60

df_worlds['Stream length'] = df_worlds['Stream length'].apply(change_hours)
df_worlds
```

```python
# Converted stream start time to different date format and sorted by date from 2015-2018
df_worlds['Stream start time'] = pd.to_datetime(df_worlds['Stream start time'])
df_worlds = df_worlds.sort_values(['Stream start time'])
df_worlds = df_worlds.reset_index(drop = True)
```

```python
# exporting cleaned data to csv
df_worlds.to_csv('worlds.csv', index = None, header = True)
```
