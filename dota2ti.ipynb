```python
#Modifying values
dota2ti = dota2ti.rename(columns={"Stream start time":"Date"})
dota2ti["Avg viewers"] = dota2ti["Avg viewers"] * dota2ti["Stream length"]
dota2ti = dota2ti.groupby(["Date"]).sum()
dota2ti["Avg viewers"] = dota2ti["Avg viewers"]/dota2ti["Stream length"]
dota2ti["Stream length"] = dota2ti["Stream length"]/60
dota2ti["Views per hour"] = dota2ti["Views"]/dota2ti["Stream length"]
dota2ti["Followers per hour"] = dota2ti["Followers gained"]/dota2ti["Stream length"]
dota2ti = dota2ti.drop(columns="Unnamed: 0")
dota2ti = dota2ti.reset_index()
dota2ti = pd.merge(dota2data, dota2ti, on="Date")
dota2ti
```
